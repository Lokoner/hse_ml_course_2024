```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../data/marketing_campaign.csv", sep=";")
df
```

```python
print('Количество жалоб: ', df['Complain'].sum())
```

```python
df["PurMnt"] = df["MntWines"] +  df["MntFruits"] + df["MntMeatProducts"] + df["MntFishProducts"] + df["MntSweetProducts"] + df["MntGoldProds"]
```

```python
print('Средний возраст людей, которые совершили покупку:', 2024 - df[df['PurMnt'] > 0]['Year_Birth'].mean())
```

```python
print('Средний возраст людей в датасете:', 2024 - df['Year_Birth'].mean())
```

Оба значения равны

# MEDIUM

```python
df['Age'] = 2024 - df['Year_Birth']
df.head()
```

```python
df['Education'] = df['Education'].replace({
    'Basic': 0,
    'Graduation': 1,
    'Master': 2,
    '2n Cycle': 3,
    'PhD': 4
})

df['Marital_Status'] = df['Marital_Status'].replace({
    'Single': 0,
    'Together': 1,
    'Married': 2,
    'Divorced': 3,
    'Widow': 4,
    'Alone': 5,
    'Absurd': 6,
    'YOLO': 7
})
```

```python
new_df = df[["Education", "Marital_Status", "Income", "Kidhome", 
                "Teenhome", "Recency", "NumWebPurchases", "NumStorePurchases", 
                "Age"]]
new_df
```

```python
sns.pairplot(new_df)
plt.show()
```

Выводы: У людей, у которых 2 ребёнка, доход ниже, чем с 0 или 1 ребенком;
        Нет семей, где есть 2 ребёнка (kid) и 2 подростка (teen) одновременно;
        Люди, у которых есть семья, доход выше, чем у тех, у кого её нет.
        Люди c наименьшими и наибольшими доходами чаще закупаются через интернет
        
Выбросы: на графиках заметны выбросы с Age (120+ лет), NumWebPurchases
         и Income (в последнем случае мб просто богатый, но удобнее считать, что выброс)
